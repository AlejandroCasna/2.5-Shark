```python
def get_detailed_location_type(location):
    if pd.notnull(location):  # Verifica si el valor no es nulo
        keywords = ['beach', 'river', 'ocean', 'lagoon', 'port']
        for keyword in keywords:
            if re.search(rf'\b{keyword}\b', location, re.IGNORECASE):
                return keyword
        if isinstance(location, str) and 'cable' in location.lower():
            return 'cable'
    return 'other'
```
